### import libraries

```python
import pandas as pd
import re
from hazm import Normalizer
```

### load dataset

```python
df = pd.read_csv("../data/farsi_dari_toxic_comments.csv")
```

### normalize Farsi text (HAZM)

```python
normalizer = Normalizer()
df["comment"] = df["comment"].apply(normalizer.normalize)
```


### 1. REMOVE EMOJIS

```python
emoji_pattern = re.compile(
    "["
    "\U0001F600-\U0001F64F"
    "\U0001F300-\U0001F5FF"
    "\U0001F680-\U0001F6FF"
    "\U0001F1E0-\U0001F1FF"
    "\U00002700-\U000027BF"
    "\U000024C2-\U0001F251"
    "]+",
    flags=re.UNICODE
)

df["comment"] = df["comment"].astype(str).apply(
    lambda x: emoji_pattern.sub("", x)
)
```


### 2. REMOVE DIGITS


```python
digit_pattern = re.compile(r"\d+")

df["comment"] = df["comment"].apply(
    lambda x: digit_pattern.sub("", x)
)
```


### 3. KEEP ONLY FARSI CHARACTERS


```python
fa_pattern = re.compile(r"[^\u0600-\u06FF\s]")

df["comment"] = df["comment"].apply(
    lambda x: fa_pattern.sub(" ", x)
)
```


### 4. CLEAN EXTRA SPACES


```python
df["comment"] = df["comment"].apply(
    lambda x: re.sub(r"\s+", " ", x).strip()
)
```


### 5. CHECK EMPTY COMMENTS


```python
print("Empty comments:", (df["comment"].str.strip() == "").sum())
```

### 6. REMOVE EMPTY + DUPLICATES


```python
df = df[df["comment"].str.strip() != ""]
df = df.drop_duplicates(subset=["comment"]).reset_index(drop=True)
```

### 7. FINAL STATS


```python
print("Remaining rows:", len(df))
print(df.head())
print(df["label"].value_counts())
```

### 8. SAVE CLEAN DATASET


```python
df.to_csv("../data/cleaned_dataset.csv", index=False)

print("Dataset saved successfully.")
```
