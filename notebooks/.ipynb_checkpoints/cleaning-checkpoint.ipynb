### cleaning my dataset

```python
import re

emoji_pattern = re.compile(
    "["
    "\U0001F600-\U0001F64F"
    "\U0001F300-\U0001F5FF"
    "\U0001F680-\U0001F6FF"
    "\U0001F1E0-\U0001F1FF"
    "\U00002700-\U000027BF"
    "\U000024C2-\U0001F251"
    "]+",
    flags=re.UNICODE
)

cleaned_texts = []

for i, text in enumerate(df['comment']):
    original_text = str(text)

    # find emojis
    emojis_found = emoji_pattern.findall(original_text)

    # remove emojis
    cleaned_text = emoji_pattern.sub("", original_text)

    # print only if emojis exist
    if emojis_found:
        print(f"Row {i} -> Removed emojis: {emojis_found}")

    cleaned_texts.append(cleaned_text)

# replace column safely
df['comment'] = cleaned_texts
```

```python
df.head()
```

```python
import re

fa_pattern = re.compile(r"[^\u0600-\u06FF\s]")

cleaned_texts = []

for i, text in enumerate(df['comment']):
    original_text = str(text)

    # keep only Farsi/Arabic letters
    cleaned = fa_pattern.sub(" ", original_text)

    # normalize spaces
    cleaned = re.sub(r"\s+", " ", cleaned).strip()

    # debug print (only if something changed)
    if original_text != cleaned:
        # print(f"\nRow {i} changed")
        print("BEFORE:", original_text)
        print("AFTER :", cleaned)

    cleaned_texts.append(cleaned)

df['comment'] = cleaned_texts
```

```python
print(df['comment'].apply(lambda x: str(x).strip() == "").sum())
```

```python
df = df[df['comment'].astype(str).str.strip() != ""]
df = df.drop_duplicates(subset=['comment'])
```

```python
df = df.reset_index(drop=True)
```

```python
print("Remaining rows:", len(df))
print(df.head())
print(df['label'].value_counts())
```
